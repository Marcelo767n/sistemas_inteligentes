```python
import pandas as pd

# Conjunto de datos: conocimiento de sintomas y enfermedades
data = {
    "Sintoma": ["fiebre", "tos", "dolor de cabeza", "fatiga"], #x
    "Posible_Enfermedad": ["gripe", "resfriado", "migrana", "anemia"] #y
}

df = pd.DataFrame(data)
print(df)
```

```python
# Consultar el conocimiento: que enfermedad puede estar relacionada con "tos"?
resultado = df[df["Sintoma"] == "tos"]
print("Resultado de consulta:\n", resultado)
```

```python
# Ver la estructura de los datos
print(df.info())

# Resumen estadistico de las variables numericas
print(df.describe())

# Ver las columnas disponibles
print("Columnas de dataset:", df.columns.tolist())
```
